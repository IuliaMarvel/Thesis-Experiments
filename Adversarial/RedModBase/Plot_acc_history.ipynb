```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_cases = ['base','reduce_easy','add_hard','add_random','reduce_random']
```

```python
dict_name = 'history_'

history = dict()

for scheme in data_cases:
    history[scheme] = np.load(dict_name+scheme+'.npy',allow_pickle='TRUE').item()[scheme]
```

```python
keys = history['base'].keys()
n_epochs = 20
```

```python
fig,axes = plt.subplots(1,3,figsize=(14,6))

for index,key in enumerate(keys):
  for scheme in data_cases:
    axes[index].plot(range(n_epochs),history[scheme][key],label=str(scheme))
    axes[index].set_xlabel('epoch')
    axes[index].set_ylabel('accuracy')
    axes[index].legend()

plt.legend()
```

```python
plt.savefig('acc_history(RedModBase).png')
```
